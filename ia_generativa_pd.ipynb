## Questão 1: Explique os seguintes conceitos fundamentais dos LLMs, fornecendo exemplos práticos e diagramas onde for relevante:
<br>

**a) Pre-training**: é o treinamento da LLM para alguma tarefa de previsão de token (por exemplo, para um modelo de geração de texto ou um modelo de classificação) com um corpus massivo de dados não rotulados. É chamado de pré-treinamento porque, embora seja uma fase de treinamento, o conhecimento codificado pelo modelo durante esta fase é fundamental para qualquer tarefa de linguagem natural subsequente. A figura a seguir mostra o processo de treinamento de LLMs.

pre_training_llm.PNG

Figura 1 - Processo de pré-treinamento e *fine-tunning* da LLM
<br>
<br>

**b) Transfer Learning**: é o processo de treinar um LLM para uma determinada tarefa e depois adaptá-lo (realizar um *fine-tunning*) para que ele realize uma nova tarefa. Por exemplo, utilizar o modelo ChatGPT, treinado pela OpenAI, e realizar um *fine-tunning* para que ele seja um assistente jurídico, ajudando na elaboração de pareceres e petições. Isso permite aproveitar o conhecimento adquirido pelo modelo para realizar a tarefa original. Em termos de arquitetura, o processo envolve a divisão do modelo em um "corpo" e uma "cabeça", onde a "cabeça" é uma rede específica para a tarefa. Durante o treinamento, os pesos do corpo aprendem características gerais do domínio de origem e esses pesos são usados para inicializar um novo modelo para a nova tarefa. Em comparação com o aprendizado supervisionado tradicional, essa abordagem normalmente produz modelos de alta qualidade que podem ser treinados com muito mais eficiência em uma variedade de tarefas subsequentes e com muito menos dados rotulados. A próxima figura compara os modelos com aprendizado supervisionado tradicional (lado esquerdo) com os modelos obtidos por *transfer learning* (lado direito).

transfer_learning.PNG

Figura 2 - Comparativo entre treinamento supervisionado tradicional (lado esquerdo) e os modelos obtidos por *transfer learning* (lado direito)
<br>
<br>

**c) Embeddings**: o uso de *embeddings* envolve representar as palavras ou sentenças em um espaço multidimensional. A distância matemática entre diferentes instâncias neste espaço representa a sua similaridade em termos de significado. Por exemplo, as palavras "rainha", "mulher", "rei" e "homem". No espaço multidimensional, onde palavras são vetores, se a representação está correta, a representação seria a seguinte:
$$
\begin{array}{ccc}
\vec{rei} + (\vec{mulher} - \vec{homem}) \approx \vec{rainha}
\end{array}
$$

Isto significa que a distância entre "mulher" e "homem" deveria ser igual à distância entre "rainha" e "rei".
<br>
<br>

**d) Transformers**: nos modelos de LLM, cada palavra de uma sentença, como "O gato sentou no tapete", é convertida em sua representação vetorial numérica. Assim, "gato" se torna uma série de números, assim como "sentou", "no" e "tapete". Os *transformers* são uma arquitetura que utiliza estes vetores numéricos, que representam palavras, e compreende os seus relacionamentos, em termos de estrutura (sintaxe) e significado (semântica). Para fazer isso, eles executam operações matemáticas para entender o relacionamento entre as palavras, produzindo novos vetores com riqueza de informações de contexto. Uma das características marcantes dos *transformers* é a capacidade de compreender os significados contextuais e matizados das palavras. O mecanismo de *self-attention* nos *transformers* permite que cada palavra em uma frase observe todas as outras palavras para entender melhor seu contexto.
Seria como se cada palavra votasse na importância de outras palavras para o seu significado. Ao considerar a frase inteira, os *transformers* podem determinar com mais precisão o papel e o significado de cada palavra, tornando suas interpretações mais ricas em contexto. Há muitos tipos de arquitetura de *transformers*. A figura a seguir mostra as arquiteturas BERT e GPT.

transformers.PNG


Figura 3 - Arquiteturas BERT e GPT de *transformers*

<br>

**e) Attention**: os modelos tradicionais de Processamento de Linguagem Natural (NLP, na sigle em inglês) processam o texto sequencialmente, o que limita o seu entendimento da estrutura da linguagem, especialmente em textos longos. O *attention* é um mecanismo que permite aos modelos relacionar diretamente palavras distantes umas às outras, independentemente de suas posições no texto. O *attention* funciona da seguinte forma: em vez de produzir um único estado oculto para a sequência de entrada, o codificador emite um estado oculto em cada etapa que o decodificador pode acessar. No entanto, usar todos os estados simultaneamente criaria uma entrada enorme para o decodificador, portanto, é necessário algum mecanismo para priorizar quais estados usar. É aqui que entra o *attention*: ele permite que o decodificador atribua uma quantidade diferente de peso, ou "attention", a cada um dos estados do codificador em cada etapa de decodificação. Esse processo é ilustrado na figura a seguir, onde o papel do *attention* é mostrado para prever o terceiro token na sequência de saída.

attention.PNG

Figura 4 - Arquitetura de *encoder-decoder* com um mecanismo de *attention* para um par de redes neurais recorrentes (RNN)
<br>
<br>

**f) Fine-Tunning**: é o processo de utilizar um modelo que foi pré-treinado para um tarefa de forma mais ampla, como geração de texto, e ajustá-lo para uma tarefa específica, como por exemplo, geração de pareceres jurídicos. Este processo envolve o mecanismo de *transfer learning*, no qual os conhecimentos obtidos no pré-treinamento são transferidos para o modelo que realizará a tarefa específica. Para realizar o *fine-tunning*, é necessário fornecer ao modelo uma amostra (de tamanho menor que a amostra usada no pré-treinamento) de dados rotulados contendo exemplos da tarefa específica que se deseja que o modelo realize. # Questão 4: Analise os seguintes prompts e identifique por que eles poderiam gerar respostas insatisfatórias ou irrelevantes:

Subquestões:

1.   Reformule cada prompt utilizando técnicas de engenharia de prompts para torná-los mais específicos e direcionados.
2.   Explique as melhorias feitas em cada caso e os motivos por trás das reformulações.



Resposta:

Os prompts citados nos exemplos 1 e 2 são muito genéricos. Ao pedir que o modelo escreva sobre cachorros, ela pode escrever muitas informações sobre esse assunto, como o comportamento, as diferentes raças, o surgimento, hábitos alimentas. Desta forma, é grande a chance do modelo fornecer informações que não são do interesse do usuário.

O mesmo princípio vale para o prompt 2 "Explique física". A física é uma ciência que se abre em vários campos de estudo, como física óptica, eletromagnetismo, mecânica dos fluídos, entre outras. O prompt solicitado é muito abrangente, sendo necessário que o usuário especifique melhor a sua necessidade.

Para melhorar os prompts, pode ser usada a técnica de engenharia chamada *Least-To-Most*. O exemplo 1 ficaria da seguinte forma:

"Escreva sobre cachorros. Para atingir esse objetivo, escreva sobre as seguintes informações:


*   Quais são os alimentos permitidos para eles? É necessário alimentá-los apenas com ração?
*   Qual a frequência ideal para dar banho neles? Quais produtos utilizar?
*   Qual a frequência ideal para levá-los ao veterinário?
*   Qual a frequência ideal para levá-los para passear?
*   Quais raças são melhores para serem criadas em apartamento?
*   Como ensiná-los a fazer as necessidades na caixa de areia?"

As melhorias executadas neste prompt se referem à torná-lo mais específico, solicitando as informações que realmente interessam ao usuário, reduzindo o risco do modelo gerar alucinações e/ou informações inúteis.

No caso do exemplo 2, a reengenharia ficaria da seguinte forma:

"Explique os seguintes tópicos sobre a ciência 'física':

*   O que é a física?
*   Como a física surgiu?
*   Qual os seus principais campos de atuação?
*   Quais os principais estudiosos que contribuíram para física e quais foram as suas contribuições?
*   Quais as principais descobertas da física nos últimos 10 anos?
*   Quais universidades têm realizado mais contribuições para a física nos últimos 10 anos?"

Novamente, a reengenharia deste prompt envolveu torná-lo mais específico, indicando exatamente os assuntos de interesse, o que pode aumentar a chance do modelo gerar informações que atendam às expectativas do usuário.

# Questão 6: Aplique a técnica de Chain of Thought (CoT) para melhorar o prompt "Explique como funciona a energia solar.", detalhando o raciocínio necessário para que o modelo forneça uma resposta completa e coerente. Explique como a aplicação da técnica CoT melhora a resposta do modelo.

Resposta:

Ao aplicar a técnica Chain of Thought (CoT), o prompt ficaria da seguinte forma:

"Explique como a energia solar funciona, detalhando passo a passo o processo, iniciando com a absorção da luz solar pelas células fotovoltaicas, a conversão da luz em energia elétrica, o papel dos inversores e a distribuição de energia. De forma complementar, discuta os fatores de eficiência, os avanços tecnológicos e os benefícios ambientais da energia solar."

A técnica de CoT melhora o prompt por meio das seguintes ações:

*   **Divisão da atividade em subatividades**: ao invés de um prompt vago, a CoT define que o prompt deve ser dividido em subatividades, a fim de guiar o modelo pelo caminho desejando. No caso do prompt em questão, foi criada a sequência lógica: absorção - conversão - distribuição;
*   **Encoraja respostas aprofundadas**: ao questionar sobre "eficiência" e "avanços tecnológicos", o modelo pode gerar respostas com maior profundidade;
*   **Expansão do contexto**: a inclusão de "benefícios ambientais" expande o contexto da resposta;
*   **Previne simplificação**: ao se solicitar informações sobre "inversores" e "distribuição", o modelo é forçado a gerar resultados mais técnicos e menos superficiais; e
*   **Melhora a coerência**: a forma como os assuntos vão progredindo no prompt ajuda a gerar respostas mais coerentes, com encadeamento lógico de conceitos.
